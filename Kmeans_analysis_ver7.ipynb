```python
# 외부 온도: 4도씩 범주화
bins_temp = np.arange(-9, 40, 4)
labels_temp = [f'{i}-{i+4}' for i in bins_temp[:-1]]
df['외부온도범위'] = pd.cut(df['외부 온도'], bins=bins_temp, labels=labels_temp, right=True, include_lowest=True)

# 차량 속도: 50km/h씩 범주화
bins_speed = [0, 50, 100, float('inf')]  # 0-50, 50-100, 100 이상
labels_speed = ['0-50', '50-100', '100+']  # 레이블 설정
df['차량속도범위'] = pd.cut(df['차량 속도'], bins=bins_speed, labels=labels_speed, right=False, include_lowest=True)

# 일사량: 0.4씩 범주화
bins_solar = np.arange(0, 1.24 + 0.4, 0.4)
labels_solar = [f'{i:.1f}-{i+0.4:.1f}' for i in bins_solar[:-1]]
df['일사량범위'] = pd.cut(df['일사량'], bins=bins_solar, labels=labels_solar, right=False, include_lowest=True)

# 습도: 20씩 범주화
bins_humidity = np.arange(0, 101, 20)
labels_humidity = [f'{i}-{i+20}' for i in bins_humidity[:-1]]
df['습도범위'] = pd.cut(df['습도'], bins=bins_humidity, labels=labels_humidity, right=False, include_lowest=True)
```

```python
# 선택된 열만 범주화할 수 있는 함수
def categorize_columns(df, interval=4, selected_columns=None):
    categorized_df = df.copy()
    
    # 선택된 열이 없으면 모든 열을 선택
    if selected_columns is None:
        selected_columns = df.columns
    
    for col in selected_columns:
        if col in df.columns:  # 열이 데이터프레임에 존재하는지 확인
            min_val = df[col].min()
            max_val = df[col].max()
            bins = np.round(np.arange(min_val, max_val + interval, interval), 1)
            labels = [f'{i}-{i+interval}' for i in bins[:-1]]
            categorized_df[col + '_범위'] = pd.cut(df[col], bins=bins, labels=labels, right=False, include_lowest=True)
    
    return categorized_df

# 특정 열만 범주화
selected_columns = ['운전석 온도', '보조석 온도', '뒷자석 온도(중앙)', '앞 루프 중앙', '앞 루프 좌측', '앞 루프 우측', '뒷 루프 좌측', '뒷 루프 우측']
df_categorized = categorize_columns(df_temp, interval=4, selected_columns=selected_columns)
```

```python
# 결합할 열 리스트
columns_to_combine = ['운전석 온도', '보조석 온도', '뒷자석 온도(중앙)', '앞 루프 중앙', '앞 루프 좌측', '앞 루프 우측', '뒷 루프 좌측', '뒷 루프 우측']

# 각 열에 대해 원본 열과 범위 열을 결합
combined_df_list = []

for column in columns_to_combine:
    combined_df = pd.concat([df, df_categorized[[column, f'{column}_범위']]], axis=1)
    combined_df_list.append(combined_df)

# 결합된 데이터프레임 확인
for combined_df in combined_df_list:
    combined_df.info()
```

```python
# 필요한 순서로 열 정렬 및 교차 테이블 생성
for combined_df, column in zip(combined_df_list, columns_to_combine):
    new_order = ['에어컨 압축기 전력 소비량', '히터의 전력 소비량', '히트 펌프', '외부 온도', f'{column}', '설정 온도', '바람세기', '일사량', 
                 '습도', '차량 속도', '내부 온도', '외부온도범위', '차량속도범위', '일사량범위', '습도범위', f'{column}_범위']
    
    print(new_order)
    combined_df = combined_df[new_order]
    
    cross_tab = pd.crosstab(index=[combined_df['외부온도범위'], combined_df[f'{column}_범위'], combined_df['차량속도범위'], combined_df['일사량범위']], 
                            columns=combined_df['습도범위'])
    
    # 교차 테이블 결과를 엑셀 파일로 저장
    excel_filename = f'/home/miserver10/workspace_vscode/K-means/result_xlsx/cross_tab_{column}.xlsx'
    cross_tab.to_excel(excel_filename)
    
    print(f"교차 테이블 결과가 '{excel_filename}' 파일로 저장되었습니다.")
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd

# 스케일러 정의
scaler = StandardScaler()

# 각 column에 대한 category_data 딕셔너리 생성
category_data_dict = {column: {} for column in columns_to_combine}

# 각 column별로 데이터 처리 및 저장
for combined_df, column in zip(combined_df_list, columns_to_combine):
    temp_ranges = combined_df['외부온도범위'].unique()
    temp_col_ranges = combined_df[f'{column}_범위'].unique()
    speed_ranges = combined_df['차량속도범위'].unique()
    solar_ranges = combined_df['일사량범위'].unique()
    humidity_ranges = combined_df['습도범위'].unique()

    # 카테고리 조합에 따른 데이터프레임 생성
    for temp_range in temp_ranges:
        for temp_col_range in temp_col_ranges:
            for speed_range in speed_ranges:
                for solar_range in solar_ranges:
                    for humidity_range in humidity_ranges:
                        # 각 카테고리 조합에 해당하는 데이터프레임 생성
                        subset = combined_df[
                            (combined_df['외부온도범위'] == temp_range) &
                            (combined_df[f'{column}_범위'] == temp_col_range) &
                            (combined_df['차량속도범위'] == speed_range) &
                            (combined_df['일사량범위'] == solar_range) &
                            (combined_df['습도범위'] == humidity_range)]
                        subset = subset.drop(['외부온도범위', f'{column}_범위', '차량속도범위', '일사량범위', '습도범위'], axis=1)  # 카테고리 이름 열 제외
                        # 열을 drop하거나 재 정렬할 때, 의도치 않게 순서가 바뀔 수 있어 재 정렬하기
                        subset = subset[['에어컨 압축기 전력 소비량', '히터의 전력 소비량', '히트 펌프', '외부 온도', f'{column}', '설정 온도', '바람세기', '일사량', '습도', '차량 속도', '내부 온도']]

                        if len(subset) >= 30:  # 데이터가 있는지 확인
                            key = f'{column}_{temp_col_range}_temp_{temp_range}_speed_{speed_range}_solar_{solar_range}_humidity_{humidity_range}'
                            category_data_dict[column][key] = subset
```

```python
# 엑셀 파일 생성
for column, category_data in category_data_dict.items():
    excel_file_path = f'/home/miserver10/workspace_vscode/K-means/result_xlsx/clustering_results_{column}.xlsx'
    with pd.ExcelWriter(excel_file_path) as writer:
        # sheet_index = 1
        for key, value in category_data.items():
            # features 설정
            temp_column = key.split('_')[0]
            features = ['외부 온도', temp_column, '차량 속도', '일사량', '습도']
            selected_features = value[features]
            scaled_value = scaler.fit_transform(selected_features)

            # KMeans 적용
            kmeans = KMeans(n_clusters=3, random_state=42)
            clusters = kmeans.fit_predict(scaled_value)

            # 클러스터링 결과를 데이터프레임에 추가
            clustered_value = value.copy()
            clustered_value['클러스터'] = clusters

            # 클러스터별 평균 계산 (스케일링 되기 전의 값으로)
            cluster_means = clustered_value.groupby('클러스터').mean()
            cluster_means = cluster_means.round(1)

            # # 고유 시트 이름 생성
            # sheet_name = f'Sheet{sheet_index}'
            # sheet_index += 1

            # 시트 이름을 외부온도범위로 설정
            temp_range = key.split('_')[3]
            sheet_name = f'temp_{temp_range}' 

            # key를 포함한 데이터프레임 생성
            key_df = pd.DataFrame({'key': [key]})
            result_df = pd.concat([cluster_means, key_df], axis = 1)

            # 결과를 엑셀 파일의 시트로 저장
            result_df.to_excel(writer, sheet_name=sheet_name, index=False)

    print(f"클러스터링 결과가 '{excel_file_path}' 파일에 저장되었습니다.")
```
