```python
#bins_temp = [-9, -5, -1, 3, 7, 11, 15, 19, 23, 27, 31, 35, 39]  # 외부 온도의 범위 설정
#labels_temp = ['-9-(-5)도', '-5-(-1)도', '31-34도']  # 각 범위의 레이블 설정

# 외부 온도를 4도씩 묶기 위한 범위 설정
bins_temp = list(range(-9, 40, 4))  # -9도에서 39도까지 4도 간격으로
labels_temp = [f'{bins_temp[i]}-{bins_temp[i+1]}' for i in range(len(bins_temp)-1)]  # 레이블 설정
```

```python
df['외부온도범위'] = pd.cut(df['외부 온도'], bins=bins_temp, labels=labels_temp, right=False)

# pd.cut의 파라미터 right = True(기본값)일 경우, include_lowest=True 필요 : 왼쪽 값이 열린 구간이 되면서 첫 번째 경계 값이 포함되지 않게 되는것을 막아준다 
# right = Flase 일때는 오른쪽 경계값이 열린 구간이 되어 문제 해결
```

```python
bins_speed = [0, 50, 100, float('inf')]  # 0-50, 50-100, 100 이상
labels_speed = ['0-50', '50-100', '100+']  # 레이블 설정
```

```python
df['차량속도범위'] = pd.cut(df['차량 속도'], bins=bins_speed, labels=labels_speed, right=False)
```

```python
# 카테고리별 데이터 저장을 위한 딕셔너리 생성
category_data = {}

# 외부 온도 범위와 차량 속도 범위에 따라 데이터를 나누고 저장
for temp_range in df['외부온도범위'].unique():
    for speed_range in df['차량속도범위'].unique():
        subset = df[(df['외부온도범위'] == temp_range) & (df['차량속도범위'] == speed_range)]
        subset = subset.drop(['외부온도범위', '차량속도범위'], axis=1)  # '외부온도범위', '차량속도범위' 열 제외
        if not subset.empty:  # 데이터가 있는지 확인
            category_data[f'temp_{temp_range}_speed_{speed_range}'] = subset

# 딕셔너리에 저장된 데이터프레임 출력
for key, value in category_data.items():
    print(f"카테고리: {key}")
    print(value)
    print()  
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 스케일러 정의
scaler = StandardScaler()

# 결과를 저장할 엑셀 파일 이름
excel_file = 'clustered_results(K,3).xlsx'

# 엑셀 파일 생성
with pd.ExcelWriter(excel_file) as writer:
    for key, value in category_data.items():
        # 선택한 특성에 대해 스케일링 수행
        scaled_value = value.copy()  # 원본 데이터 복사
        scaled_value = scaler.fit_transform(scaled_value)

        # KMeans 적용
        kmeans = KMeans(n_clusters=3, random_state=42)
        clusters = kmeans.fit_predict(scaled_value)

        # 클러스터링 결과를 데이터프레임에 추가
        clustered_value = value.copy()
        clustered_value['클러스터'] = clusters

        # 클러스터별 평균 계산 (스케일링 되기 전의 값으로)
        cluster_means = clustered_value.groupby('클러스터').mean()

        # 결과를 엑셀 파일의 시트로 저장
        cluster_means.to_excel(writer, sheet_name=key)

# 엑셀 파일 저장 완료 메시지 출력
print(f"클러스터링 결과가 '{excel_file}' 파일에 저장되었습니다.")
```
