### Data scaling

```python
X=df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

### Elbow method

```python
from sklearn.cluster import KMeans

#  WCSS (Within-Cluster Sum of Squares), inertia ì™€ ê°™ì€ ê°œë…
wcss = []
for i in range(1,9):    
    kmeans = KMeans(n_clusters = i,init = "k-means++",random_state = 0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
print(wcss)
```

```python
plt.plot(range(1,9),wcss,marker = '*')
plt.xticks(range(1,9))
plt.show()
```

### silhouette_score

```python
from sklearn.metrics import silhouette_score

# ë°ì´í„°í”„ë ˆìž„ê³¼ X ì„¤ì •
# df = ... (ë°ì´í„°í”„ë ˆìž„ ì„¤ì •)
# X = df[['feature1', 'feature2', ...]].values  # í´ëŸ¬ìŠ¤í„°ë§í•  íŠ¹ì§• ì„ íƒ

# ì‹¤ë£¨ì—£ ê³„ìˆ˜ë¥¼ ì €ìž¥í•  ë¦¬ìŠ¤íŠ¸ ì´ˆê¸°í™”
silhouette_scores = []

# í´ëŸ¬ìŠ¤í„° ìˆ˜ 2ë¶€í„° 8ê¹Œì§€ì— ëŒ€í•´ ì‹¤ë£¨ì—£ ê³„ìˆ˜ë¥¼ ê³„ì‚°
for i in range(2, 9):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=0)
    kmeans.fit(X_scaled)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# ê²°ê³¼ ì¶œë ¥
print(silhouette_scores)

# ì‹¤ë£¨ì—£ ê³„ìˆ˜ ê·¸ëž˜í”„ ê·¸ë¦¬ê¸°
plt.plot(range(2, 9), silhouette_scores, marker='o')
plt.title('Silhouette Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()

# ìµœì ì˜ í´ëŸ¬ìŠ¤í„° ìˆ˜ ì°¾ê¸°
optimal_k = np.argmax(silhouette_scores) + 2  # range ì‹œìž‘ì´ 2ì´ë¯€ë¡œ ì¸ë±ìŠ¤ì— 2ë¥¼ ë”í•¨
print(f"Optimal number of clusters based on silhouette score: {optimal_k}")
```

### k-means & ì‹œê°í™”

```python
kmeans = KMeans(n_clusters=3, init="k-means++", random_state=0)
kmeans.fit(X_scaled)
labels = kmeans.labels_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', alpha=0.6)
centroids = kmeans.cluster_centers_
centroids_pca = pca.transform(centroids)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], marker='x', s=300, c='red', edgecolors='k', label='Centroids')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-means Clustering (Standardized)')
plt.legend()
plt.show()
```
