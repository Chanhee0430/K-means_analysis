###  library import

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
X = df
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#  X 
# df = ... ( )
# X = df[['feature1', 'feature2', ...]].values  #    

# WCSS     
wcss = []
silhouette_scores = []

#   1 8   K-Means 
for i in range(1, 9):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=0)
    kmeans.fit(X_scaled)

    # WCSS  
    wcss.append(kmeans.inertia_)

    #   2    
    if i > 1 : 
        cluster_labels = kmeans.labels_
        silhouette_avg = silhouette_score(X_scaled, cluster_labels)
        silhouette_scores.append(silhouette_avg)

# WCSS     
print("WCSS:", wcss)
plt.plot(range(1, 9), wcss, marker='*')
plt.title('WCSS Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 9))
plt.show()

#       
print("Silhouette Scores:", silhouette_scores)
plt.plot(range(2, 9), silhouette_scores, marker='o')
plt.title('Silhouette Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 9))
plt.show()

#    
optimal_k = np.argmax(silhouette_scores) + 2  # range  2  2 
print(f"Optimal number of clusters based on silhouette score: {optimal_k}")
```

```python
kmeans = KMeans(n_clusters=3, init="k-means++", random_state=0)
kmeans.fit(X_scaled)
labels = kmeans.labels_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# labels  CPU 
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', alpha=0.6)
centroids = kmeans.cluster_centers_
centroids_pca = pca.transform(centroids)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], marker='x', s=300, c='red', edgecolors='k', label='Centroids')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-means Clustering (Standardized)')
plt.legend()
plt.show()
```

```python
df_scaled = pd.DataFrame(X_scaled, columns=df.columns)

df_scaled
```

```python
df_scaled['Cluster'] = labels

cluster_summary = df_scaled.groupby('Cluster').mean()
print("Cluster Summary (Mean Values):")
print(cluster_summary)
```
